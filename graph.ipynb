```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import pandas as pd
datacsv = pd.read_csv('istherecorrelation.csv')

x = datacsv['NL Beer consumption [x1000 hectoliter]']
y = datacsv['WO [x1000]']

plt.plot(x, y, 'o')
plt.xlabel('NL Beer consumption [x1000 hectoliter]')
plt.ylabel('WO [x1000]')
plt.title('NL Beer consumption [x1000 hectoliter] vs WO x1000')

# Fit with polyfit
fitted = linregress(x, y)


plt.plot(x, fitted[1] + fitted[0] * x, '-')
plt.show()
print("pearsen correlation coefficient is: {}".format(fitted[2]))
plt.savefig('output.png', dpi=300)
```
